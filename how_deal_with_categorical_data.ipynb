```python
#onehotencoder
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
d={"fruits":["Apple","Banana","Orange","Apple","Banana","Orange"],"color":["Red","yellow","orange","Red","yellow","orange"],
  "size":[12,23,14,12,23,14]}
d
```

```python
df=pd.DataFrame(d)
df
```

```python
encoder=OneHotEncoder()
```

```python
#fit -use formula but not use calculated value by formula
#fit_transform -use formula and use calculated value
encoder.fit_transform(df[["fruits"x=]]) #3 unique rows and add these as column so (6,3)
```

```python
encoder.fit_transform(df[["fruits"]]).toarray() 
```

```python
en=encoder.fit_transform(df[["fruits"]])
```

```python
en
```

```python
x=pd.DataFrame(en.toarray(), columns=encoder.get_feature_names_out())
```

```python
x
```

```python
pd.concat([df,x],axis=1)
```

```python
en=encoder.fit_transform(df[["fruits","color"]]).toarray()
```

```python
x=pd.DataFrame(en, columns=encoder.get_feature_names_out())
```

```python
x
```

```python
pd.concat([df,x],axis=1)
```
