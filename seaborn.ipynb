```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
distribution plot
-distplot
-joinplot
-pairplot
```

```python
df=sns.load_dataset("tips") #if we use pandas we write read_csv("tips.csv")
df.head()
```

```python
#find correlation-When data is numeric
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True))
```

```python
sns.heatmap(df.corr(numeric_only=True),annot=True) #also display valuesn
```

```python
sns.heatmap(df.corr(numeric_only=True),annot=True,fmt=".3f") #fmt shows decimal as desired .3- 3 decimal places
```

```python
jointplot- ait allows to study relationship b/w 2 numeric values the central chart displays their relationship 
```

```python
sns.jointplot(x='tip', y='total_bill',data=df,kind='hex')
```

```python
#it shows best  fit line
sns.jointplot(x='tip', y='total_bill',data=df,kind='reg')
```

```python
pairplot-scatterplot. in which one variable in the same dta row is matched  
Plot pairwise relationships in a dataset.

By default, this function will create a grid of Axes such that each numeric
variable in ``data`` will by shared across the y-axes across a single row and
the x-axes across a single column. The diagonal plots are treated
differently: a univariate distribution plot is drawn to show the marginal
distribution of the data in each column.

It is also possible to show a subset of variables or plot different
variables on the rows and columns.

This is a high-level interface for :class:`PairGrid` that is intended to
make it easy to draw a few common styles. You should use :class:`PairGrid`
directly if you need more flexibility.
```

```python
sns.pairplot(df,hue=None)
```

```python
sns.pairplot(df,hue="sex") # it differentiate b/w maleand female
```

```python
#distplot to check distribution of columns
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```
