```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([2,5,8])
y=np.array([0,25,8])
plt.plot(x,y)
plt.show()
```

```python
#highlight point m=marker
plt.plot(x,marker="o")
plt.show()
```

```python
#tp colour line color=c
plt.plot(x,marker="o",color="r")
plt.show()
```

```python
#marker edge color mec
plt.plot(x,marker="o",color="r",mec="g")
plt.show()
```

```python
plt.plot(x,marker="h",color="r",mec="g")
plt.show()
```

```python
plt.plot(x,marker="H",color="r",mec="g")
plt.show()
```

```python
#size of marker inc
plt.plot(x,marker="H",color="r",mec="g",markersize=10)
plt.show()
```

```python
#line style
plt.plot(x,marker="o",color="r",mec="g",markersize=10,linestyle="dashed")
plt.show()
```

```python
#line style
plt.plot(x,marker="o",color="r",mec="g",markersize=10,linestyle="--")
plt.show()
```

```python
#linewidth
plt.plot(x,marker="o",color="r",mec="g",markersize=10,linestyle="--",linewidth=5)
plt.show()
```

```python
#marker face colour mfc
plt.plot(x,marker="h",color="r",mec="g",markersize=10,linestyle="--",linewidth=5,mfc="purple")
x1=np.array([2,6,5,3])
y1=np.array([4,5,8,9])
plt.plot(x1,y1,marker="h",color="yellow",mec="b",markersize=10,linestyle="solid",linewidth=5,mfc="black")
plt.show()
```

```python
#grid,label,title
plt.plot(x,y,marker="h",color="r",mec="g",markersize=10,linestyle="--",linewidth=5,mfc="purple")
plt.xlabel("students")
plt.ylabel("study hours")
plt.title("university students vs study hours")
plt.grid(axis="x")
plt.show()
```

```python
#simple scatter plot for one variable
a=np.array([33,45,23,87,98,12,13])
b=np.array([1,2,34,56,78,90,65])
plt.scatter(a,b,color="green")
```

```python
#two variables relation
a=np.array([33,45,23,87,98,12,13])
b=np.array([1,2,34,56,78,90,65])
plt.scatter(a,b,color="purple")
a1=np.array([31,32,65,56,78,87,98])
b1=np.array([21,22,23,43,38,87,23])
plt.scatter(a1,b1,color="violet")
```

```python
#colour array
colors=["green","black","yellow","red","orange","pink","brown"]
a=np.array([33,45,23,87,98,12,13])
b=np.array([1,2,34,56,78,90,65])
plt.scatter(a,b,color="purple")
a1=np.array([31,32,65,56,78,87,98])
b1=np.array([21,22,23,43,38,87,23])
plt.scatter(a1,b1,c=colors)
```

```python
#color through numbers
colors=np.array([12,23,13,14,20,23,17])
a=np.array([33,45,23,87,98,12,13])
b=np.array([1,2,34,56,78,90,65])
plt.scatter(a,b,c=colors)
a1=np.array([31,32,65,56,78,87,98])
b1=np.array([21,22,23,43,38,87,23])
plt.scatter(a1,b1,c=colors)
```

```python
#MAP COLORS
colors=np.array([12,23,13,14,20,23,17])
a=np.array([33,45,23,87,98,12,13])
b=np.array([1,2,34,56,78,90,65])
plt.scatter(a,b,c=colors)
a1=np.array([31,32,65,56,78,87,98])
b1=np.array([21,22,23,43,38,87,23])
plt.scatter(a1,b1,c=colors,cmap="viridis")

```
