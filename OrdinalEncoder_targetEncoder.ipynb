```python
#ordinal encoder, Label encoder 
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder

d1={"Education":["Phd","Pg","Ug","School"],
    "Review":["Excellent","Very good","Good","poor"],
    "Result":["Yes","No","Yes","No"]}
df1=pd.DataFrame(d1)
df1
```

```python
ode= OrdinalEncoder(categories=[["School","Ug","Pg","Phd"],
                                ["Good","Very good","Excellent","poor"]])
```

```python
x=["Education","Review"]
```

```python
df1[x]=ode.fit_transform(df1[x])
```

```python
print(df1)
```

```python
#result- label output
le=LabelEncoder()
df1["Result"]=le.fit_transform(df1["Result"])
df1
```
