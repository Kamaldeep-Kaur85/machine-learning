```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Salary_Data.csv")
```

```python
df.head(3)
```

```python
df.info()
```

```python
x=df[["YearsExperience"]]
```

```python
y=df["Salary"]
```

```python
x
```

```python
y
```

```python
#division
from sklearn.model_selection  import train_test_split
```

```python
x_train, x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train) #fitting best fit line  and logical model ready to predict 
```

```python
lr.intercept_ #theta not
```

```python
lr.coef_ #theta 1
```

```python
plt.scatter(x,y)
```

```python
y_pred=lr.predict(x_test)
```

```python
x_pred=lr.predict(x_train)
```

```python
plt.scatter(x_test,y_test,color="red",marker="+")
plt.scatter(x_train,y_train,color="orange",marker="*")
plt.scatter(x_test,y_pred,color="green")
plt.plot(x_test,y_pred,color="green")
```

```python
plt.scatter(x_test,y_test,color="green")    # green points actual points
plt.plot(x_train,x_pred,color="red")  
plt.title("salary vs experience")
plt.xlabel("experience")
plt.ylabel("Salary")
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
r2=r2_score(y_test,y_pred)
```

```python
r2
```

```python
mae=mean_absolute_error(y_test,y_pred)
mae
```

```python
from sklearn.metrics import mean_squared_error
#mean squared error
mse=mean_squared_error(y_test,y_pred)
mse
```

```python
#root mean squared error
rmse=mse**(1/2)
```

```python
rmse
```

```python
from sklearn.metrics import root_mean_squared_error
#root mean squared error
mse=root_mean_squared_error(y_test,y_pred)
mse
```
