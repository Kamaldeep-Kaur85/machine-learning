```python
#fillna
import pandas as pd
import numpy as np
```

```python
d={"Apple":[45,25,np.nan, 15, np.nan,54],"Orange":[12,13,56,78,95,np.nan],"Grapes":[1,2,3,4,8,9],"papaya":[45,np.nan,74,56,23,15]}
```

```python
d
```

```python
df=pd.DataFrame(d)
df
```

```python
#to check null values
df.isnull()
```

```python
#which column have nan and no. of nan values in column
df.isnull().sum()
```

```python
df.notnull()
```

```python
df.notnull().sum()
```

```python
df.fillna(0)
```

```python
df.fillna(1) #but this method is not efficient it can effect accuracy
```

```python
df.fillna(method="pad") #it will show warning 
```

```python
df.fillna(method="ffill") 
```

```python
df.ffill() #if first column has nan it remains nan
```

```python
df.bfill()
```
